### Read an image(which contains different objects) using matplotlib and check the shape of the image?
#### Display the image?
#### Identify the different objects in the image? 
#### Use slicing to get the objects and display them?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=plt.imread("image.jpg")
print(img.shape)
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
```

```python
plt.imshow(img[:,:,0],cmap='gray')
```

```python
plt.imshow(img[:,0:1000])
```

```python
plt.imshow(img[:400,:1000])
```

```python
plt.imshow(img[100:300,600:900])
```

```python
plt.imshow(img[:100,:1000])
```
